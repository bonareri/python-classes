## Challenge: Build a Movie Search Application (Using an API)

### Objective  
Create a command-line application that allows users to search for movies and retrieve detailed information using a public movie API.

### Requirements

#### 1. API Access  
Use a public movie API such as:

- OMDb API  
- TMDb API  

*(OMDb is recommended for beginners because it is simpler to use.)*

#### 2. User Input  
- Prompt the user to enter a movie title.

#### 3. API Request  
- Make a **GET request** to the API.
- Pass the movie title as a query parameter.
- Include your API key in the request.

#### 4. Data Handling  
Parse the JSON response and extract relevant information such as:

- Title  
- Year  
- Genre  
- Director  
- IMDb Rating  
- Plot  

#### 5. Output  
Display the movie details in a clean and readable format.

Example output:

Title: Inception
Year: 2010
Genre: Action, Sci-Fi
Director: Christopher Nolan
IMDb Rating: 8.8
Plot: A thief who steals corporate secrets...

#### 6. Error Handling  
Handle cases such as:

- Movie not found  
- Empty input  
- Invalid API key  
- Network errors  

## Bonus Tasks

### 1. Multiple Results
If multiple movies match the search:
- Display a numbered list
- Let the user choose one to view full details

### 2. Save to File
Allow the user to:
- Save the movie details to a `.csv` file

### 3. Search by Year
Allow users to search using both:

- Movie title  
- Year  

Example:

Enter movie name: Avatar
Enter year: 2009

### 4. Top Rated Movies Feature
Allow users to:
- Search by keyword (e.g., "Batman")
- Display movies sorted by IMDb rating (if multiple results)
