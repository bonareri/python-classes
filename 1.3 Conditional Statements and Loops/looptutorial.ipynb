```python
import random

secret_num = random.randint(1, 100)

attempts = 3
count = 0

print("I have generate a number between 1 and 100")
print(f"You have {attempts} attempts to guess the number")

while count < attempts:
    guess_num = int(input("Guess a number between 1 and 100"))
    count += 1

    if guess_num > secret_num:
        print("Too high")
    elif guess_num < secret_num:
        print("Too low")
    else:
        print("Congratulations, you guesses right")

if count == attempts and guess_num!= secret_num:
    print(f"You are out of attempts, the secret number was {secret_num}")
```
